```python
import numpy as np
import pandas as pd 
import cv2
import os
import math
```

```python
path = '/Kodlar/MachineLearning/Proje2/data'  
data_0 = pd.DataFrame()
data_1 = pd.DataFrame()
data_2 = pd.DataFrame()
data_0['name']   =  os.listdir(path+'/0') 
data_1['name']   =  os.listdir(path+'/1')
data_2['name']   =  os.listdir(path+'/2')  

```

```python

paths_0 = []
paths_1 = []
paths_2 = []

class_0 = []
class_1 = []
class_2 = []
for image in data_0['name']:
    class_0.append([0,0,1])
    paths_0.append(path+'/0'+'/'+image)
for image in data_1['name']:
    class_1.append([0,1,0])
    paths_1.append(path+'/1'+'/'+image)
for image in data_2['name']:
    class_2.append([1,0,0])
    paths_2.append(path+'/2'+'/'+image)

data_0['path']  = paths_0 
data_0['class'] = class_0 
data_1['path'] = paths_1
data_1['class'] = class_1  
data_2['path'] = paths_2 
data_2['class'] = class_2 
```

```python
data_ = pd.DataFrame() 
data_ = data_.append(data_0)
data_ = data_.append(data_1) 
data_ = data_.append(data_2) 
data_

```

```python
data_ = data_.sample(frac=1).reset_index(drop=True)
data_
```

```python
train_images = []
# train verileri için resimleri dönüştürme
for file in data_["path"]:
    img = cv2.imread(file) 
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray = cv2.resize(gray, (10,10))
    gray = gray.reshape(gray.shape[0]*gray.shape[1])

    train_images.append(gray) 
```
